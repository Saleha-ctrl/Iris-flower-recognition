```python
import joblib 

model = joblib.load("Iris_recognition_model.pkl")
scale = joblib.load("scaler.pkl")

new_data = [[5.1, 3.5, 1.4, 0.2]]

scaled_data = scale.transform(new_data)
predict = model.predict(scaled_data)

if (predict == 0):
    print("Predicted flower is setosa")

elif (predict == 1):
    print("Predicted flower is versicolor")

elif (predict == 2):
    print("Predicted flower is virginica")
```
